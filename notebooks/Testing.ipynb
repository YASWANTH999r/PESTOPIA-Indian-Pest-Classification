```python
import os
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
```

```python
model = load_model('model\Custommodel.h5')

def load_and_preprocess_image(img_path):
    img = image.load_img(img_path, target_size=(150, 150))  
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)  
    img_array /= 255.0  
    return img_array
```

```python
test_images_dir = r"C:\Dataset\Data\Locust"

for img_file in os.listdir(test_images_dir):
    img_path = os.path.join(test_images_dir, img_file)
    img = load_and_preprocess_image(img_path)
    predictions = model.predict(img)
    predicted_class = np.argmax(predictions, axis=1)
    class_names = ['locust', 'Pieris_canidia', 'large_cutworm', 'red_spider', 'yellow_rice_borer']
    predicted_label = class_names[predicted_class[0]]
    
    print(f'{img_file} -> Predicted Class: {predicted_label}')
```
